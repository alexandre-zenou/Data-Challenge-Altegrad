```python
import pickle
from collections import Counter, defaultdict
import numpy as np
import pandas as pd
import torch

def load_pkl(path):
    with open(path, "rb") as f:
        return pickle.load(f)

data_path = "data_baseline/data/"
obj = load_pkl(data_path+ f"train_graphs.pkl")
print(type(obj))

from torch_geometric.data import Data

if isinstance(obj, Data):
    print("Single Data")
elif isinstance(obj, list):
    print("List length:", len(obj), " | first:", type(obj[0]))
elif isinstance(obj, dict):
    print("Dict keys:", obj.keys())
else:
    print("Unknown container")
```
